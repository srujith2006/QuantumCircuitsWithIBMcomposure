```python
from qiskit import QuantumCircuit
from qiskit.qasm2 import loads
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
# OpenQASM code
qasm_code = """
OPENQASM 2.0;
include "qelib1.inc";

qreg q[3];
x q[0];
x q[1];
h q[1];
z q[2];
"""

# Load circuit from QASM
qc = loads(qasm_code)

# Draw circuit
print(qc.draw())

plt.show()

# Generate statevector
qi = Statevector.from_instruction(qc)

# Print statevector
print("Statevector:")
print(qi)

# Get probabilities
prob = qi.probabilities_dict()

plot_histogram(prob)
```
