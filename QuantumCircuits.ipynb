```python
import qiskit
from qiskit import QuantumCircuit
```

```python
qc=QuantumCircuit(2)
qc.x(0)
qc.h(1)

qc.draw('mpl')
```

```python
from qiskit import quantum_info as qi
psi_0=qi.Statevector(qc)
psi_0.draw('latex')
psi_1=qi.Statevector(qc)
psi_1.draw('latex')
```

```python

psi_0.draw('latex')
```

```python
from IPython import display
output1=psi_0.draw('latex')
output2=psi_1.draw('latex')
display.display(output1,output2)
psi_0.draw('bloch')
```

```python
psi_1.draw('bloch')
```

```python
qc=QuantumCircuit(5)
qi.Statevector(qc).draw('bloch')
```

```python
qc.x(0)
qc.rx(30*3.1/180,1)
qc.y(2)
qc.ry(45*3.1/180,3)
qc.z(4)
qc.draw('mpl')
```

```python
qi.Statevector(qc).draw('bloch')
```

```python
qc2=QuantumCircuit(2)
qc2.cx(0,1)
qc2.draw('mpl')
```

```python
q=qi.Operator(qc2)
q.draw('latex')
```

```python
import numpy as np
psi_0=qi.Statevector([1,1,0,0]/np.sqrt(2))
psi_0.draw('latex')
```

```python
qc=QuantumCircuit(2)
qc.initialize(psi_0,[0,1])
qc.cx(0,1)
qc.draw('mpl')
```
